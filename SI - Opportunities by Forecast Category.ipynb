```python
import pandas as pd
import os
import numpy as np
import datetime as dt
from os import listdir
from os.path import isfile, join
from directories import forecast_cat_w_op_file
```

```python
#-----------------------------------------------------------------------------------------------------------------------------
# IT_Forecast_Category_w_Opp
#-----------------------------------------------------------------------------------------------------------------------------
df = pd.read_csv(forecast_cat_w_op_file, encoding='latin-1', sep=";")

for col in df.columns:
    if "Date" in col:
        df[col] = df[col].astype(str)
        df[col] = pd.to_datetime(df[col], format="%d/%m/%Y")
    else:
        pass

columns = [c for c in df.columns if not "Currency" in c]
df = df.loc[:,columns]

num = []

for col in df.columns:
    if "Amount" in col:
        num.append(col)
    else:
        pass
    
df[num] = df[num].apply(lambda x: x.str.replace(",",".", regex=False)).astype(float)
```

```python
df["Rete"] = np.where(
    df["Owner Role"].str.contains("IT Agency and Reseller"), 
    "Indiretta", 
    "Diretta")
```

```python
df["Created Date M"] = df["Created Date"].dt.to_period("M")
df["Created Date Q"] = df["Created Date"].dt.to_period("Q")
df["Created Date H"] = np.where(
    df["Created Date Q"].astype(str).str.contains("Q1|Q2"),
    "H1",
    "H2")
df["Created Date Y"] = df["Created Date"].dt.to_period("Y")


df["Close Date M"] = df["Close Date"].dt.to_period("M")
df["Close Date Q"] = df["Close Date"].dt.to_period("Q")
df["Close Date H"] = np.where(
    df["Close Date Q"].astype(str).str.contains("Q1|Q2"),
    "H1",
    "H2")
df["Close Date Y"] = df["Close Date"].dt.to_period("Y")
```

```python
df = df.loc[df["Close Date"] >= "2024-01-01",:].copy()
#.to_excel(r"C:\Users\Raffaele.Sportiello\OneDrive - Wolters Kluwer\Documents\Presentazioni\Sales\Salesforce adoption\20240116_SFDC_fcst_,category_dates_v1.xlsx")
```

```python

df_pivoted_amount = df.pivot_table(
    index=[
        df["Close Date Q"],
        pd.Grouper(freq="M", key="Close Date", how="sum"), 
        "Forecast Category"], 
    columns=[
        df["Created Date Y"],
        df["Created Date Q"],
        pd.Grouper(freq="Q", key="Created Date", how="sum")],
    values="Incremental Amount",
    aggfunc="sum")
#.to_excel(r"C:\Users\Raffaele.Sportiello\OneDrive - Wolters Kluwer\Documents\Presentazioni\Sales\Salesforce adoption\prova.xlsx")
```
