```python
import pandas as pd

# Caricamento dei file CSV
file_abbo = r'C:\Users\Raffaele.Sportiello\OneDrive - Wolters Kluwer\Documents\AI\Z_DBTAX_ABBO.csv'
file_prod =  r'C:\Users\Raffaele.Sportiello\OneDrive - Wolters Kluwer\Documents\AI\Z_DBTAX_MD_ZCODPROD.csv'

# Lettura dei file con separatore punto e virgola
df_abbo = pd.read_csv(file_abbo, sep=';')
df_prod = pd.read_csv(file_prod, sep=';')
```

```python
# Funzione per rinominare le colonne mantenendo solo la parte tra virgolette
def rename_columns(df):
    new_column_names = []
    for col in df.columns:
        # Trova la prima occorrenza di testo tra virgolette
        start = col.find('"')
        end = col.find('"', start + 1)
        if start != -1 and end != -1:
            new_column_names.append(col[start+1:end])
        else:
            new_column_names.append(col)
    df.columns = new_column_names

# Rinominare le colonne per entrambe le tabelle
rename_columns(df_abbo)
rename_columns(df_prod)
```

```python
# Conversione delle colonne che contengono "Data" in tipo data
for col in df_abbo.columns:
    if "Data" in col:
        df_abbo[col] = pd.to_datetime(df_abbo[col], format='%Y%m%d', errors='coerce')

# Conversione delle colonne "Valore PV New" e "N. Abb." in tipo numerico
df_abbo['Valore PV New'] = pd.to_numeric(df_abbo['Valore PV New'], errors='coerce')
df_abbo['N. Abb.'] = pd.to_numeric(df_abbo['N. Abb.'], errors='coerce')
```

```python
import pandas as pd

# Caricamento dei file CSV
file = r'C:\Users\Raffaele.Sportiello\OneDrive - Wolters Kluwer\Documents\Base dati\Budest\Budest dettaglio ordine\Z_DBTAX_BUDEST_APAC.csv'

# Lettura dei file con separatore punto e virgola
df = pd.read_csv(file, sep=';', low_memory=False)


# Funzione per rinominare le colonne mantenendo solo la parte tra virgolette
def rename_columns(df):
    new_column_names = []
    for col in df.columns:
        # Trova la prima occorrenza di testo tra virgolette
        start = col.find('"')
        end = col.find('"', start + 1)
        if start != -1 and end != -1:
            new_column_names.append(col[start+1:end])
        else:
            new_column_names.append(col)
    df.columns = new_column_names

# Rinominare le colonne per entrambe le tabelle
rename_columns(df)
```
