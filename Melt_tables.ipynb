```python
import pandas as pd
from pandas import DataFrame
import calendar
```

```python
# Funzione per pivotare i dati e aggiungere i codici di lookup
def reshape_and_add_lookup_updated(table, lookup, subco_column_name):
    # Rinomina la prima colonna in 'Subco' per coerenza
    table = table.rename(columns={subco_column_name: "Subco"})

    # Trasformazione del DataFrame per avere i mesi in riga
    reshaped = table.melt(id_vars=["Subco"], var_name="Month", value_name="Budget")

    # Aggiunta di una colonna con i codici di lookup
    reshaped['Codice'] = reshaped['Subco'].map(lookup)

    # Riordino delle colonne
    reshaped = reshaped[["Subco", "Codice", "Month", "Budget"]]

    # Conversione della colonna 'Month' in un formato data che rappresenta il primo giorno di ogni mese
    reshaped['Month'] = reshaped['Month'].apply(lambda x: pd.to_datetime('2024-' + str(list(calendar.month_abbr).index(x[:3])) + '-01'))

    return reshaped
```

```python
# Tabella di lookup
lookup_table = {
    "Enterprise": 22, 
    "Large Account": 17, 
    "Center South": 8, 
    "North East": 9, 
    "North West": 10, 
    "Inside": 23, 
    "Indirect - IPSOA": 3, 
    "Indirect - Resellers": 2
}
```

```python
# Percorso del file Excel
file_path = r'C:\Users\Raffaele.Sportiello\OneDrive - Wolters Kluwer\Documents\AI\Quota24.xlsx'

# Lettura delle due tabelle dal file Excel
first_table_range = pd.read_excel(file_path, sheet_name='Quota Mensilizzata', skiprows=0, nrows=9, usecols='A:M')
second_table_range = pd.read_excel(file_path, sheet_name='Quota Mensilizzata', skiprows=13, nrows=9, usecols='A:M')
```

```python
# Trasformazione delle tabelle e aggiunta dei codici di lookup
reshaped_first_table = reshape_and_add_lookup_updated(first_table_range, lookup_table, "Subco")
reshaped_second_table = reshape_and_add_lookup_updated(second_table_range, lookup_table, "Service")
```
