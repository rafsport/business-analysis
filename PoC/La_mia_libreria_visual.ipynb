### Matplotlib

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.plot([5, 8, 1, 4, 4],[4, 5, 9, 9, 2], "c");
```

```python
x = [40, 56, 91, 105]
y = [121, 184, 121, 149]
```

```python
plt.plot(x, y, 
         linewidth=1.0, 
         ls="-.", 
         marker="p", 
         markersize=10, 
         markerfacecolor="w");
```

```python
plt.plot(x, y, "mD:");
```

```python
plt.plot(x, y, color="y")
plt.title("Titolo del grafico", color="blue")
plt.xlabel("Dati sull'asse X", color=(0.1, 0.2, 0.5))
plt.ylabel("Dati sull'asse Y", color="#5A6FC2")
plt.grid(True)
plt.legend(["Legenda"], loc=2);
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 8, 15], "b*-")
plt.plot([1, 3, 5, 7], [5, 4, 8, 10], "g^-.")
plt.plot([2, 4, 6, 8], [2, 5, 4, 12], "ro:")
plt.legend(["Dato Blu", "Dato Verde", "Dato Rosso"], loc="best")
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 8, 15], "b*-")
plt.subplot(1, 2, 2)
plt.plot([1, 3, 5, 7], [5, 4, 8, 10], "g^-.");
```

```python
plt.subplot(3, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 8, 15], "b*-")
plt.subplot(3, 2, 2)
plt.plot([1, 3, 5, 7], [5, 4, 8, 10], "g^-.")
plt.subplot(3, 2, 3)
plt.plot([1, 4, 6, 8], [2, 5, 4, 12], "ro:")
plt.subplot(3, 2, 4)
plt.plot([1, 2, 3, 5], [2, 7, 4, 11], "mD--")
plt.subplot(3, 2, 5)
plt.plot([1, 4, 6, 8], [2, 5, 4, 12], "ro:")
plt.subplot(3, 2, 6)
plt.plot([1, 2, 3, 5], [2, 7, 4, 11], "mD--")
```

### Seaborn
<a id='Seaborn'></a>

```python
import seaborn as sns
```

```python
churn_data = pd.read_csv("Datasets/file3_telco_customer_churn.csv")
```

```python
churn_data.dtypes
```

```python
sns.countplot(data=churn_data, x="gender")
```

```python
sns.barplot(data=churn_data, x="gender", y="MonthlyCharges")
```

```python
sns.barplot(data=churn_data, 
            x="Churn", 
            y="MonthlyCharges", 
            palette="rainbow")
```

```python
sns.boxplot(churn_data.MonthlyCharges);
```

```python
sns.boxplot(churn_data.MonthlyCharges, orient="v");
```

```python
sns.distplot(churn_data.tenure[:50], 
             hist=False, kde=True, rug=True)
```

```python
sns.violinplot(data=churn_data, x="Churn", y="tenure", 
               palette="Set1")
```

```python
sns.stripplot(data=churn_data, x="MonthlyCharges", y="TechSupport", 
              jitter=True);
```

```python
anscombe = sns.load_dataset("anscombe")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.sample(5)
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", 
           col="time",
           hue="smoker", 
           row="sex");
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip")
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg")
```

```python
sns.stripplot(data=tips, x="day", y="total_bill", jitter=True)
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill")
```
