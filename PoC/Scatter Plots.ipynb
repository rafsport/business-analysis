```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Scatterplot dove tutti i punti hanno gli stessi colori e dimensioni

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

plt.scatter(x,y, s=100, c="green", marker="o", edgecolor="black", alpha=0.75)

plt.tight_layout()

plt.show()

# Matplotlib Marker Styles: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
```

## Scatterplot con colori e dimensioni differenti

```python
plt.style.use('seaborn')

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
size = [700, 500, 900, 700, 500, 700, 200, 500, 300, 700, 100, 200, 800, 100, 900, 200, 500, 600, 700, 500]

plt.scatter(x,y, s=size, c=colors, cmap="Greens", marker="o", edgecolor="black", alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Intensit√†')

plt.tight_layout()

plt.show()

# Matplotlib Colormaps: https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
```

## Scatterplot su scala logaritmica

```python
plt.style.use('seaborn')

data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/07-ScatterPlots/2019-05-31-data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, c=ratio, cmap='summer',
            edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```
