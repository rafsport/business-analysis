```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

## Plot di valori nel tempo

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle="solid")
plt.tight_layout()

plt.show()
```

## Plot di valori nel tempo: formatting dell'asse

```python
plt.style.use('seaborn')

dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle="solid")

plt.gcf().autofmt_xdate() # Sistema automaticamente la posizione delle date; #gcf --> "get current figure"

date_format = mpl_dates.DateFormatter("%b, %d %y")  # Classe del modulo mpl_dates alla quale passo il formato di data desiderato
plt.gca().xaxis.set_major_formatter(date_format) #gca --> "get current axis"

plt.tight_layout()

plt.show()
```

## Plot di valori nel tempo: trasformazione di stringhe in date

```python
plt.style.use('seaborn')

data = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/08-TimeSeries/data.csv')

data['Date'] = pd.to_datetime(data['Date']) #trasformo la colonna di date nel formato Data
data.sort_values('Date', inplace=True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.title('Bitcoin Prices')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.tight_layout()

plt.show()
```
