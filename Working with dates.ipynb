# Tipi di date

```python
import pandas as pd
import numpy as np
```

```python
data = np.array([[1, 2], [3, 4]])
begin_date = "2022-01-01"
df = pd.DataFrame({'date': pd.date_range(begin_date, periods=10, freq="W")})
```

```python
df["date_strftime"] = df["date"].dt.strftime('%B')
```

```python
df["date_datetime64[M]"] = df["date"].astype('datetime64[M]')
```

```python
df["date_datetime64[Y]"] = df["date"].astype('datetime64[Y]')
```

```python
import datetime as dt

df["date_dt.date"] = df["date"].apply(lambda x: dt.date(x.year,x.month,1)) 
df["date_dt.date_values_astype_datetime"] = df["date"].apply(lambda x: dt.date(x.year,x.month,1)) .values.astype('datetime64[D]')
```

# Busyday count

```python
import calendar
```

```python
holidays=["2021-01-01","2021-01-06","2021-04-05","2021-05-01","2021-06-02","2021-08-15","2021-11-01","2021-12-08","2021-12-25","2021-12-26"]

df['N. gg lavorativi'] = np.busday_count(df["date"].apply(lambda x: (pd.to_datetime(dt.date(x.year,x.month,1))) ).values.astype('datetime64[D]'), 
                                             (df['date']+pd.Timedelta(days=1)).values.astype('datetime64[D]'), 
                                             weekmask=[1,1,1,1,1,0,0], holidays=holidays)
```

```python
df['N. gg lavorativi nel mese'] = np.busday_count(
    df["date"].apply(
        lambda x: 
        (pd.to_datetime(dt.date(x.year,x.month,1))) ).values.astype('datetime64[D]'),
    
    (df['date'].apply(
        lambda d: 
        dt.date(d.year, d.month, calendar.monthrange(d.year, d.month)[-1]))).values.astype('datetime64[D]'),
    
    weekmask=[1,1,1,1,1,0,0], holidays=holidays)
```
