```python
unione_comuni.replace(regex={"'": " "}, inplace=True)
unione_comuni.replace(regex={r'\s+': " ", r'^(-|_|\s|null|altro)*$': np.NaN}, inplace=True)
```

#### Segnalare quando un comune appartiene ad una Unione

Creazione del datframe "unione_comuni_lista_comuni": identifica quando un comune appartiene ad una Unione.  
La categorizzazione avviene su base Comune ("City"), per cui tutti gli enti che nella tabella "companies" hanno associato un Comune che è presente in questo dataframe avranno il campo "Unione A Cui Appartiene Comune" valorizzato con quella Unione di Comuni.

```python
unione_comuni_lista_comuni = unione_comuni[["Comuni Appartenenti All'Unione", "Descrizione Unione Di Comuni"]].apply(lambda x: x.str.title())
unione_comuni_lista_comuni = unione_comuni_lista_comuni[unione_comuni_lista_comuni["Comuni Appartenenti All'Unione"] !=  "Totale Unione"]
unione_comuni_lista_comuni.columns = ["City", "Unione A Cui Appartiene Comune"]
```

#### Descrivere le unioni di comuni

Creazione del datframe "unione_comuni" che permette di valorizzare i campi "N. Comuni Appartenenti A Unione", "Comuni Appartenenti A Unione" e "N. Residenti Unione Comuni" nella tabella "companies".  
L'associazione avviene sulla base del campo "Descrizione Unione Di Comuni" adeguatemente ripulito per ottimizzare l'unione.

```python
unione_comuni["Comuni Appartenenti All'Unione"] = unione_comuni.groupby("Descrizione Unione Di Comuni")["Comuni Appartenenti All'Unione"].transform(lambda x: ", ".join(x)).replace(regex={"Totale Unione, ":""}).str.title()
```

```python
unione_comuni = unione_comuni.loc[unione_comuni["R."] == 0, ["Descrizione Unione Di Comuni", "Nr. Comuni", "Comuni Appartenenti All'Unione", "Pop. Totale"]]
unione_comuni.rename(columns={"Pop. Totale":"N. Residenti Unione Comuni", "Comuni Appartenenti All'Unione": "Comuni Appartenenti A Unione", "Nr. Comuni": "N. Comuni Appartenenti A Unione"}, inplace=True)
```

```python
unione_comuni["Descrizione Unione Di Comuni"] = unione_comuni["Descrizione Unione Di Comuni"].str.lower().replace(regex={r'(\s*(de|da|ne|su)ll\')': " ",
                                                                                                                         r'(\s*de[loagie\']{1,3})\s': " ",
                                                                                                                         r'(\s*da[ileaog\']{1,3})\s': " ", 
                                                                                                                         r'(\s*ne[ileaog\']{1,3})\s': " ", 
                                                                                                                         r'(\s*su[ileaog\']{1,3})\s': " ", 
                                                                                                                         r'(\s*d(i|\')*\s)': " ", 
                                                                                                                         r'comuni(t)*(à|a\')*': "", 
                                                                                                                         "unione": "",
                                                                                                                         "unite' des communes valdotaines": "",
                                                                                                                         "\?": ""}).replace(regex={r'\'': ""}).replace(regex={r'\s{2,}': " "}).replace(regex={r'^\s+': ""})
```

* Sostituzione nel campo "Mail1" delle email "pec" con email dal campo "Mail2","Mail3","Mail4","Mail5", quando presenti
* Arricchimento del campo "Mail1" con le altre email  
* Applicazione del formato "lower" (tutto minuscolo) al campo "Mail1"

```python
companies.loc[(companies["Mail2"].notna()) & (~companies["Mail2"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False,na=False)) & (companies["Mail2"] != companies["Mail1"]) & (companies["Mail1"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False)), "Mail1"]  =  companies["Mail2"]
companies.loc[(companies["Mail3"].notna()) & (~companies["Mail3"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False,na=False)) & (companies["Mail3"] != companies["Mail1"]) & (companies["Mail1"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False)), "Mail1"]  =  companies["Mail3"]
companies.loc[(companies["Mail4"].notna()) & (~companies["Mail4"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False,na=False)) & (companies["Mail4"] != companies["Mail1"]) & (companies["Mail1"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False)), "Mail1"]  =  companies["Mail4"]
companies.loc[(companies["Mail5"].notna()) & (~companies["Mail5"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False,na=False)) & (companies["Mail5"] != companies["Mail1"]) & (companies["Mail1"].str.contains("pec.|legalmail.|cert.|certificata.|verimail.",case=False)), "Mail1"]  =  companies["Mail5"]

companies.loc[companies["Mail1"].isna(), "Mail1"] =  companies["Mail2"]
companies.loc[companies["Mail1"].isna(), "Mail1"] =  companies["Mail3"]
companies.loc[companies["Mail1"].isna(), "Mail1"] =  companies["Mail4"]
companies.loc[companies["Mail1"].isna(), "Mail1"] =  companies["Mail5"]

companies["Mail1"] = companies["Mail1"].str.lower()
```

Creazione della colonna "Industry" con la seguente regola:
* se "Nome_categoria" contiene le parole "Comuni e loro Consorzi e Associazioni" E "Denominazione_ente" contiene le parole "Comune di " &rarr; valorizzo con la stringa "Comune"
* se "Nome_categoria" contiene le parole "Comuni e loro Consorzi e Associazioni" E "Denominazione_ente" non contiene le parole "Comune di " &rarr; valorizzo con la stringa "Consorzio di Comuni"
* altrimenti valorizzo con il valore contenuto in "Nome_categoria"

```python
companies.loc[(companies["Nome_categoria"] == "Comuni e loro Consorzi e Associazioni") & (companies["Denominazione_ente"].str.contains("Comune ", case=False)), "Industry"] = "Comune"
companies.loc[(companies["Nome_categoria"] == "Comuni e loro Consorzi e Associazioni") & (~companies["Denominazione_ente"].str.contains("Comune ", case=False)), "Industry"] = "Consorzio di Comuni"
companies.loc[(companies["Industry"].isna(), "Industry")] = companies["Nome_categoria"]
```
